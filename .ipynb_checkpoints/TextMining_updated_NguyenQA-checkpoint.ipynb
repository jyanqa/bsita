<font color='navy'> `DftoText()` is function to convert a corpus from original dataframe into string. There are 3 outputs using this function which are **text4**, **text10**, **text500**. They are colections of the 4, 10, and 500 negative reviews from *absita* dataset. </font>

```python
def DftoText(df):  
    list = []
    string = ""   
    df.apply(lambda x: list.append(x) if isinstance(x, str) else x) 
    for aa in list:  
        aa += '. '
        string += aa     
    return string  
```

### Obtain bi-grams
In this section, `ngrams()` is a function to extract n-grams of words.
- Input: a text file and number of grams
- Output: a chart of 10 n-grams whose the highest frequences.

```python
def nbigram(corpus,n):
    sentences = nltk.sent_tokenize(corpus)
    covec = CountVectorizer(ngram_range=(n, n))
    bag_of_words = covec.fit_transform(sentences)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in covec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    grams = covec.get_feature_names()
    df_bigrams = pd.DataFrame(words_freq[:10], columns = ['Review' , 'count'])
    df_vis = df_bigrams.groupby('Review').sum()['count'].sort_values(ascending=False).iplot(kind='bar', yTitle='Count',colors='Red', linecolor='black', title='Top 10 bigrams in corpus')
    print(grams)
    return df_vis
```

### Vectorise it with bag-of words method

```python
def bag_of_words(X):
    sentences = nltk.sent_tokenize(X)
    vectorizer = CountVectorizer()
    bagofword = vectorizer.fit_transform(sentences)
    words = vectorizer.get_feature_names()
    df_bow = pd.DataFrame(bagofword.toarray(), columns = words)
    cm = sns.light_palette("pink", as_cmap=True)
    df_colored = df_bow.style.background_gradient(cmap=cm)
    return df_colored
```

### Vectorise it TF-IDF methods 

```python
def tfidf(X):
    sentences = nltk.sent_tokenize(X)
    vectorizer = TfidfVectorizer()
    tfidf = vectorizer.fit_transform(sentences)
    words = vectorizer.get_feature_names()
    df_tfidf = pd.DataFrame(tfidf.toarray(), columns = words)
    cm = sns.light_palette("pink", as_cmap=True)
    df_colored = df_tfidf.style.background_gradient(cmap=cm)
    return df_colored
```

### Form a document-term matrix

```python
def Clean(doc):
    stop=set(stopwords.words('italian'))
    exclude=set(string.punctuation)
    lemma=WordNetLemmatizer()
    stop_free=" ".join([i for i in doc.lower().split() if i not in stop])
    punc_free=''.join(ch for ch in stop_free if ch not in exclude)
    normalized=" ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized
```

```python
def DocumentTermMatrix(X,n_rows):
    doc_complete = X.iloc[0:n_rows,0].values.tolist()
    doc_clean = [Clean(doc).split() for doc in doc_complete]
    ''''''
    np.random.seed(42)
    cv = CountVectorizer(min_df=2,max_df=50,ngram_range=(1,2), token_pattern=None, tokenizer=lambda doc:doc,preprocessor=lambda doc:doc)
    words_freq = cv.fit_transform(doc_clean)
    words = np.array(cv.get_feature_names())
    df_tdm = pd.DataFrame(words_freq.toarray(), columns = words)
    ''''''
    cm = sns.light_palette("navy", as_cmap=True)
    df_colored = df_tdm.style.background_gradient(cmap=cm)
    return df_colored
    
```

```python
class CleanText( BaseEstimator, TransformerMixin ):


    def _init_( self, lang = "italian"):
        self.lang = lang
        self.stemmer = SnowballStemmer(self.lang)

    def fit( self, X, y = None ):
        return self

    def clean (self, x):
        tex = cleancorpus(x)
        tokens = tokenCorpus(tex)
        return tokens

    def transform( self, X, y = None ):
        return self.clean(X) 
```

```python
'''CLEAN'''
class CleanText( BaseEstimator, TransformerMixin ):
    
    def __init__( self, lang = "italian"):
        self.lang = lang
        self.stemmer = SnowballStemmer(self.lang)
       
    def fit( self, X, y = None ):
        return self 
    
    def clean( self, x ):
        words   = [self.stemmer.stem(word) for word in word_tokenize(x.lower()) if word.isalpha() and word not in stopwords.words("italian")]
        return " ".join(words)
    
    def transform( self, X, y = None ):
        return X.apply(self.clean)

    
'''PIPELINE''' 
from sklearn.pipeline import Pipeline
from sklearn.pipeline import FeatureUnion
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import StandardScaler

pipe = Pipeline([("extract", FeatureUnion([("terms", Pipeline([('clean', CleanText()), 
                                                               ('tfidf', TfidfVectorizer())])),
                                          ])),
                 ("select", SelectKBest(score_func = chi2)),
                 ("scale", StandardScaler(with_mean = False))]) 
```

```python
# extract features
pipe_extract = FeatureUnion([("terms", Pipeline([('clean', CleanText()), 
                                                 ('tfidf', TfidfVectorizer())])),
                             ])

# select and scale features
pipe_select_scale = Pipeline([("select", SelectKBest(score_func = chi2)),
                              ("scale", StandardScaler(with_mean = False))]) 

# extract features
# you can also use bi-grams:
X_extract = pipe_extract.set_params(terms__tfidf__ngram_range = (1,2)).fit_transform(X_train, y_train)

# extract all features
X_select_scale = pipe_select_scale.set_params(select__k = 500).fit_transform(X_extract, y_train)
#print(X_select_scale)
```

```python
'''PIPELINE'''
pipe1 = Pipeline([("extract", FeatureUnion([("terms", Pipeline([('clean', CleanText()), 
                                                               ('tfidf', TfidfVectorizer())]))])),
                 ("scale", StandardScaler(with_mean = False))]) 
```

```python
'''TRANSFORM AND REDUCE DIMENTIONS USING PRINCIPLE COMPONENTS ANALYSIS'''
X = pipe1.fit_transform(X_train[0:1000])
pca = PCA(n_components=0.90, random_state=42)
X_reduced= pca.fit_transform(X.toarray())
```

```python
'''COMPUTING DISTORTION OVER THE RANGE OF K '''
distortions = []
K = range(7,15)
for k in K:
    k_means = KMeans(n_clusters=k, random_state=42).fit(X_reduced)
    k_means.fit(X_reduced)
    distortions.append(sum(np.min(cdist(X_reduced, k_means.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
'''DEFINING LINES'''
X_line = [K[0], K[-1]]
Y_line = [distortions[0], distortions[-1]]
'''PLOT ELBOW'''
plt.plot(K, distortions, 'b-')
plt.plot(X_line, Y_line, 'coral')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show() #Distortion computes the sum of squared distances from each point to its assigned center. 
'''WE CHOOSE k = 9'''
```

```python
#spacy download it_core_news_sm
spacy.load('it_core_news_sm') 
parser = Italian()

def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens
```

```python
from nltk.corpus import wordnet as wn
def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
otherstop =  ['molto']
it_stop = set(nltk.corpus.stopwords.words('italian') + otherstop)
```

```python
def prepare_text_for_lda(text):
    tokens = tokenize(text)
    tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in it_stop]
    tokens = [get_lemma(token) for token in tokens]
    return tokens
```

```python
text_data = []
with open ('bsita.csv') as f:
    for line in f:
        tokens = prepare_text_for_lda(line)
        text_data.append(tokens)
```

```python
dictionary = corpora.Dictionary(text_data)
bow = [dictionary.doc2bow(text) for text in text_data]
```

```python
# save corpus and dictionary
pickle.dump(bow, open('bow.pkl', 'wb'))
dictionary.save('dictionary.gensim')
```

```python
# 3 topics
NUM_TOPICS = 3
ldamodel = gensim.models.ldamodel.LdaModel(bow, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
ldamodel.save('model3.gensim')
#----------
topics = ldamodel.print_topics(num_words= 4)
for topic in topics:
    print(topic)
```
